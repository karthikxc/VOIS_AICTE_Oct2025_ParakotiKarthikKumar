### About Dataset

### Context
The dataset contains detailed information about Airbnb listings, hosts, and booking activities. It provides insights into pricing, availability, host activity, and property details across different neighborhoods and property types. This dataset can be used to analyze trends in the short-term rental market, understand customer preferences, and perform exploratory data analysis for hospitality and tourism research.

### Source
This dataset was compiled from Airbnbâ€™s publicly available listing information. It includes host details, property characteristics, reviews, and availability data for different locations.

### Data fields
id - Unique identifier for each listing

NAME - Name/title of the Airbnb listing

host id - Unique identifier for the host

host_identity_verified -Whether the hosts identity is verified (yes/no)

host name - Name of the host

neighbourhood group - Major geographical area (e.g., Brooklyn, Manhattan)

neighbourhood - Specific neighborhood where the property is located

lat- Latitude of the listing

long- Longitude of the listing

country- Country of the listing

country code- ISO country code

instant_bookable- Whether the property allows instant booking

cancellation_policy- Type of cancellation policy applied

room type- Type of room offered (Entire home/apt, Private room, Shared room)

Construction year- Year the property was constructed

price- Price per night (USD)

service fee- Service fee charged

minimum nights- Minimum number of nights required to book

number of reviews- Total reviews received

last review- Date of the most recent review

reviews per month- Average number of reviews per month

review rate number- Average rating value from guests

calculated host listings count- Number of listings the host manages

availability 365-Number of available days in a year

house_rules- Special rules provided by the host

license- License information (if applicable)

```python
def remove_outlier(series):
    sorted(series)
    Q1,Q3 = series.quantile([0.25,0.75])
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return lower_bound, upper_bound
```

```python
low, high = remove_outlier(df['price'])
```

```python
df['price'] = np.where(df['price']>high,high,df['price'])
```

```python
df['price'] = np.where(df['price']<low,low,df['price'])
```
